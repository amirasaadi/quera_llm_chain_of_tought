```python
# وارد کردن کلاس Document
from langchain_core.documents import Document

# فرض کنید این متن را از یک ایمیل کپی کرده‌ایم
raw_text = "جلسه مدیران به روز دوشنبه موکول شد."

# ساختن یک Document استاندارد
# ما متن را در page_content و اطلاعات جانبی را در metadata می‌ریزیم
doc = Document(
    page_content=raw_text,
    metadata={
        "source": "email_inbox",
        "sender": "admin@company.com",
        "date": "2023-10-25",
        "priority": "high"
    }
)

#  نمایش خروجی
print(f"--- محتوای اصلی ---\n{doc.page_content}")
print(f"\n--- شناسنامه (Metadata) ---\n{doc.metadata}")
```
