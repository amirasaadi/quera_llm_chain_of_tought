```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

model = ChatOpenAI(
    base_url="https://api.tapsage.com/openai/v1", # یا سرویس انتخابی شما
    api_key=os.getenv("YOUR_API_KEY"),
    model="gpt-4o",
    temperature=0 # دمای صفر برای کارهای استدلالی بهتر است
)
```

```python
# دقت کنید: مثال شامل بخش "Reasoning" است
examples = """
نام برند: اسنپ
تحلیل:
1. کوتاه و تلفظ آسان است (یک بخشی).
2. معنای سرعت (Snap) را تداعی می‌کند.
3. با ماهیت تاکسی اینترنتی همخوانی دارد.
نتیجه: نام عالی.

نام برند: شرکت حمل و نقل سریع‌السیر برادران و شرکا
تحلیل:
1. بسیار طولانی است (۷ کلمه).
2. به خاطر سپردن آن سخت است.
3. حس قدیمی بودن می‌دهد.
نتیجه: نام ضعیف.
"""
```

```python
template = """
تو یک مشاور برندینگ هستی. برای نام برند ورودی، ابتدا مرحله به مرحله تحلیل کن و سپس نتیجه را بگو.
از الگوی زیر پیروی کن:

{examples}

نام برند: {brand_name}
تحلیل:
"""

prompt = ChatPromptTemplate.from_template(template)
```

```python
chain = prompt | model

response = chain.invoke({
    "examples": examples,
    "brand_name": "دیجی‌کالا"
})

print(response.content)
```
