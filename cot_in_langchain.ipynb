```python
import os

from dotenv import load_dotenv
from langchain_cohere import ChatCohere
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()
```

```python
model = ChatCohere(
    cohere_api_key=os.getenv("COHERE_API_KEY"),
    model="command-a-03-2025",
    temperature=0,
)
```

```python
template = """
مسئله زیر را حل کن:
{question}

لطفاً مرحله به مرحله فکر کن و تمام محاسبات را بنویس.
"""
```

```python
prompt = ChatPromptTemplate.from_template(template)

chain_zero_shot = prompt | model | StrOutputParser()
```

```python
from IPython.display import display, Markdown

answer = ""

for chunk in chain_zero_shot.stream({"question": question}):
    text = chunk.content if hasattr(chunk, "content") else str(chunk)
    answer += text
```

```python
answer = (
    answer
    .replace(r"\[", "\n$$\n")
    .replace(r"\]", "\n$$\n")
    .replace(r"\(", "$")
    .replace(r"\)", "$")
)


rtl_answer = f"""
<div dir="rtl" style="text-align: right; line-height: 2;">

{answer}

</div>
"""

display(Markdown(rtl_answer))
```
