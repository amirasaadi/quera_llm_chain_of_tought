```python
import os

from dotenv import load_dotenv
from langchain_cohere import ChatCohere
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()
```

```python
model = ChatCohere(
    cohere_api_key=os.getenv("COHERE_API_KEY"),
    model="command-a-03-2025",
    temperature=0,
)
```

```python
# دقت کنید: مثال شامل بخش "Reasoning" است
examples = """
نام برند: اسنپ
تحلیل:
1. کوتاه و تلفظ آسان است (یک بخشی).
2. معنای سرعت (Snap) را تداعی می‌کند.
3. با ماهیت تاکسی اینترنتی همخوانی دارد.
نتیجه: نام عالی.

نام برند: شرکت حمل و نقل سریع‌السیر برادران و شرکا
تحلیل:
1. بسیار طولانی است (۷ کلمه).
2. به خاطر سپردن آن سخت است.
3. حس قدیمی بودن می‌دهد.
نتیجه: نام ضعیف.
"""
```

```python
template = """
تو یک مشاور برندینگ هستی. برای نام برند ورودی، ابتدا مرحله به مرحله تحلیل کن و سپس نتیجه را بگو.
از الگوی زیر پیروی کن:

{examples}

نام برند: {brand_name}
تحلیل:
"""

prompt = ChatPromptTemplate.from_template(template)
```

```python
chain = prompt | model

response = chain.invoke({
    "examples": examples,
    "brand_name": "چیستا"
})

print(response.content)
```
